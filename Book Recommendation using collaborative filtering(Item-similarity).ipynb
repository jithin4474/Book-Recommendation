## Book recommendation using collaborative filtering

```python
import pandas as pd
```

```python
ratings=pd.read_csv('Data Sets\Ratings.csv')
col_list = ["ISBN", "Book-Title"]
books=pd.read_csv('Data Sets\Books.csv', usecols=col_list)
ratings=pd.merge(books,ratings)
ratings=ratings.head(100158)
ratings
```

```python
user_ratings=ratings.pivot_table(index=['User-ID'],columns=['Book-Title'],values='Book-Rating')
user_ratings.head()
```

```python
user_ratings=user_ratings.dropna(thresh=10,axis=1).fillna(0)
user_ratings.head()
```

```python
item_similarity_df=user_ratings.corr(method='pearson')
item_similarity_df
```

```python
def get_similar_books(user_pref):
    
    similar_books=pd.DataFrame()
    for book,rating in user_pref:
        similar_score=item_similarity_df[book]*(rating-5)
        similar_score=similar_score.sort_values(ascending=False)
        similar_books=similar_books.append(similar_score,ignore_index=True)
        
    similar_books=similar_books.sum().sort_values(ascending=False)
    
    return similar_books
```

```python
user=[]
print("You can rate book in 0-10 range")
t = input("no of books you want to rate:")
for i in range(int(t)):
    book=input("Book:")
    rating=input("rating:")
    user.append((book,int(rating)))
print(user)

similar_book=get_similar_books(user)
similar_book
```
